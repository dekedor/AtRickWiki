```python
import requests
import json
import re
import time
```

```python
def GetTextFromArticle(articleID, fulltext=False):
    """Gets the text for a wikipedia article specified by the numeric article ID.
    If fulltext is true, returns the plaintext article with sections delimited by == [SECTION] ==
    If fulltext is false, returns just the introduction section
    Wikipedia API options: 
        prop=extracts is used to request page text
        pageids uses the page_id (from page table) to identify the article
        format=json returns a json object
        explaintext returns plaintext
        exsectionformat=wiki denotes sections as == [SECTION] ==
        exintro= requests only the page introduction
    """
    try:
        baseURL = "https://en.wikipedia.org/w/api.php?action=query&prop=extracts&pageids={}&format=json&explaintext&exsectionformat=wiki{}"
        intro = "&exintro="
        if fulltext:
            intro = ""
        reqURL = baseURL.format(articleID,intro)
        r = requests.get(reqURL)
        j = json.loads(r.text)
        text = j['query']['pages'][str(articleID)]['extract']
        return text
    except:
        return None
    
```

```python
GetTextFromArticle(25, fulltext=True)
```
